# Packages

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import joblib as jb
import pydotplus as pyd
from IPython.display import Image
import pydot

```

# Reading dataset

```python
#spliting data into input and output for model
#reading file
df=pd.read_csv('music.csv')
x=df.drop(columns=['genre'])
y=df['genre']
```

# creating model for decision Tree

```python
# creating model for training
model=DecisionTreeClassifier()
model.fit(x,y)
```

# Predicting output

```python
model.predict([[21,1],[22,0]])
```

# Visualizing Decision Tree

```python
#visualizinig model
opv=tree.export_graphviz(model,feature_names=['age','gender'], class_names=sorted(y.unique()),
                         label='all',rounded=True,filled=True)
graph = pyd.graph_from_dot_data(opv)
Image(graph.create_png())
```

# creating Train and test set

```python
#check accuracy of the model
#by creating train and test dataset
model1=DecisionTreeClassifier()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model1.fit(x_train,y_train)
```

# testing results set

```python
#predicting new trained model
predicts=model1.predict(x_test)
predicts
```

```python
#checking accuracy of the new model
score=accuracy_score(y_test,predicts)
score
```
